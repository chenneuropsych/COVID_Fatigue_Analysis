```python
f_path = "HRV_features_30seconds.xlsx"
og_data = pd.read_excel(f_path, header=0)
og_data.head()
```

```python
copy = og_data
copy['Delta_fatigue'] = pd.Series()
copy['Change_from_Baseline'] = pd.Series()
```

```python
candidates = og_data['SubjID'].unique()
for candidate in candidates:
  print(candidate)
  can_df = og_data.loc[og_data['SubjID'] == candidate]
  blocks = can_df['Block'].unique()
  prev = -1
  for block in blocks:
    print(block)
    block_df = can_df.loc[can_df['Block'] == block]
    fatigue_rating = block_df['fatigue_rating'].unique()[0]
    if block == 0:
      baseline = fatigue_rating
    if prev != -1:
      delta = fatigue_rating - prev
      baseline_shift = fatigue_rating - baseline
    else:
      delta = -1
      baseline_shift = -1
    prev = fatigue_rating
    mask = (copy['SubjID'] == candidate) & (copy['Block'] == block)
    copy.loc[mask, 'Delta_fatigue'] = delta
    copy.loc[mask, 'Change_from_Baseline'] = baseline_shift

copy.to_excel("HRV_features_30seconds_withDelta.xlsx")
```

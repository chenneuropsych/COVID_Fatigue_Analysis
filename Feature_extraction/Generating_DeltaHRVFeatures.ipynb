```python
# Import necessary libraries
import numpy as np
import pandas as pd
```

```python
f_path = "HRV_features_BlockLevel_withDelta.xlsx"
og_data = pd.read_excel(f_path, header=0)
copy = og_data
og_data.head()
```

```python
cols = list(og_data.columns)
remove = ['Block','SubjID', 'Fatigue_Rating', 'Age', 'Delta_fatigue', 'Change_from_Baseline']
columns = [col for col in cols if col not in remove]
columns
```

```python
for column in columns:
  new_column = 'Delta_'+column
  copy[new_column] = pd.Series()
```

```python
candidates = og_data['SubjID'].unique()
for candidate in candidates:
  print(candidate)
  can_df = og_data.loc[og_data['SubjID'] == candidate]
  blocks = can_df['Block'].unique()
  prev = -1
  for block in blocks:
    print(block)
    block_df = can_df.loc[can_df['Block'] == block]
    for column in columns:
      print(column)
      val = block_df[column].unique()[0]
      if block == 0:
        baseline_val = val
      if prev != -1:
        delta = val - prev
      else:
        delta = 0
      prev = val
      mask = (copy['SubjID'] == candidate) & (copy['Block'] == block)
      new_column = 'Delta_'+column
      copy.loc[mask, new_column] = delta

copy.to_excel("HRV_features_BlockLevel_HRVDelta.xlsx")
```
